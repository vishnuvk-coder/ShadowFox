```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
result = classifier("I love this internship")

print(result)
```

```python
result = classifier("This project is difficult")

print(result)
```

```python
texts = [
    "vishnu is very good boy",
    "khushkant is very bad boy",
    "bhawna is very harworking girl",
    "I love my family"
]

results = classifier(texts)

for text, result in zip(texts, results):
    print("Text:", text)
    print("Sentiment:", result['label'])
    print("Score:", round(result['score'], 2))
    print()
```

This project uses Hugging Face Transformers for sentiment analysis.

The NLP language model predicts whether a sentence has positive or negative sentiment.

Strengths:
- Fast prediction
- Easy implementation
- High accuracy

Limitations:
- Can sometimes misunderstand context
- Depends on pretrained model quality
